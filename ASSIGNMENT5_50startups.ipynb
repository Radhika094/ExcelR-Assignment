```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as  sns
import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import influence_plot
```

```python
data= pd.read_csv('/content/50_Startups.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data1 = data.rename({'R&D Spend':'RDS','Administration':'ADMS','Marketing Spend':'MKTS'},axis=1)
data1.head()
```

```python
data1.corr()
```

```python
sns.pairplot(data1)
```

Model Building

```python
model = smf.ols('Profit~RDS+ADMS+MKTS',data=data1).fit()
```

```python
model.params
```

```python
model.tvalues , model.pvalues
```

```python
model.rsquared , model.rsquared_adj
```

Collinearity Check

```python
rsq1 = smf.ols('RDS~ADMS+MKTS',data=data1).fit().rsquared
vif1 = 1/(1-rsq1)
rsq2 = smf.ols('ADMS~RDS+MKTS',data=data1).fit().rsquared
vif2 = 1/(1-rsq2)
rsq3 = smf.ols('MKTS~RDS+ADMS',data=data1).fit().rsquared
vif3 = 1/(1-rsq3)

#creating a dataframe
df = {'Variable':['RDS','ADMS','MKTS'],'VIF':[vif1,vif2,vif3]}
vif_df = pd.DataFrame(df)
vif_df
```

**Conclusion : No Collinearity Found**

Residual Analysis

```python
#Q-Q Plot
sm.qqplot(model.resid,line='q')
plt.title('Q-Q Plot')
plt.show()
```

Test for Homoscedasticity or Hetroscedasticity

```python
def standard_values(vals) : return (vals-vals.mean())/vals.std()
```

```python
plt.scatter(standard_values(model.fittedvalues),standard_values(model.resid))
plt.title('Residual Plot')
plt.xlabel('Standardised Fitted Values')
plt.ylabel('Standardised Residual Value')
plt.show()
```
