***PREPROCESSING***

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
```

```python
str_col=['workclass','education','maritalstatus',
         'occupation','relationship','race','sex','native']
```

```python
for i in str_col:
  train[i]=label_encoder.fit_transform(train[i])
  test[i]=label_encoder.fit_transform(test[i])
```

```python
col_names =list(train.columns)
col_names
```

```python
train_X = train[col_names[0:13]]
train_X
```

```python
train_Y = train[col_names[13]]
train_Y
```

```python
test_x = test[col_names[0:13]]
test_x
```

```python
test_y = test[col_names[13]]
test_y
```

**Model Building**

Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
Gnbmodel=GaussianNB()
```

```python
train_pred_gau=Gnbmodel.fit(train_X,train_Y).predict(train_X)
train_pred_gau
```

```python
test_pred_gau = Gnbmodel.fit(train_X,train_Y).predict(test_x)
test_pred_gau
```

```python
train_acc_gau = np.mean(train_pred_gau==train_Y)
test_acc_gau = np.mean(test_pred_gau==test_y)
```

```python
train_acc_gau , test_acc_gau
```

Multi-Nominal Naive Bayse

```python
from sklearn.naive_bayes import MultinomialNB
Mnmodel = MultinomialNB()
```

```python
train_pred_multi=Mnmodel.fit(train_X,train_Y).predict(train_X)
train_pred_multi
```

```python
test_pred_multi=Mnmodel.fit(train_X,train_Y).predict(test_x)
test_pred_multi
```

```python
train_acc_multi = np.mean(train_pred_multi==train_Y)
test_acc_multi = np.mean(test_pred_multi==test_y)
```

```python
train_acc_multi , test_acc_multi
```
