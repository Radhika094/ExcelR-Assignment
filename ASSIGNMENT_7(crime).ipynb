HIERARCHICAL CLUSTERING

```python
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return (x)
```

```python
crime = (crime.iloc[:,1:])
crime.head()
```

```python
#Normalizing Data
crime_norm = norm_func(crime.iloc[:,1:])
crime_norm
```

```python
#Dendrogram
dendrogram = sch.dendrogram(sch.linkage(crime_norm, method='single'))
```

```python
dendrogram = sch.dendrogram(sch.linkage(crime_norm, method='average'))
```

```python
#Clusters
hc= AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='single')
```

```python
clusters = pd.DataFrame(hc.fit_predict(crime_norm),columns=['Clusters'])
clusters['Clusters'].value_counts()
```

```python
y_hc= hc.fit_predict(crime_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
crime['Clusters']=clusters
crime
```

**K-Means Clustering**

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
X = np.random.uniform(0,1,1000)
Y = np.random.uniform(0,1,1000)
df_xy = pd.DataFrame(columns=['X','Y'])
df_xy.X = X
df_xy.Y = Y
df_xy.plot(x='X',y='Y',kind='scatter')
```

```python
model1=KMeans(n_clusters=5).fit(df_xy)
df_xy.plot(x='X',y='Y',c=model1.labels_,kind='scatter',s=10,cmap=plt.cm.coolwarm)
```

```python
crime = pd.read_csv('/content/crime_data.csv')
```

```python
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return (x)

df_norm = norm_func(crime.iloc[:,1:])

df_norm.head(10)
```

```python
model = KMeans(n_clusters=5)
model.fit(df_norm)

model.labels_
```

```python
md= pd.Series(model.labels_)
crime['clust']=md
```

```python
crime.iloc[:,1:7].groupby(crime.clust).mean()
```

```python
crime.head()
```

```python
Wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,random_state=2)
  kmeans.fit(df_norm)
  Wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),Wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('Wcss')
plt.show()
```

**DBSCAN**

```python
crime = pd.read_csv('/content/crime_data.csv')
crime.head()
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
array=crime.values
array
```

```python
ss = StandardScaler().fit(array)
X = ss.transform(array)
X
```

```python
dbscan = DBSCAN(eps=0.9, min_samples=5)
dbscan.fit(X)
```

```python
dbscan.labels_ #-1=noisy samples
```

```python
c1 = pd.DataFrame(dbscan.labels_,columns=['cluster'])
c1
```

```python
pd.concat([crime,c1],axis=1)
```
